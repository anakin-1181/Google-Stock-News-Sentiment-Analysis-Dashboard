```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
fetch_url = "https://finviz.com/quote.ashx?t="
ticks = {"NVDA", "AMD", "AMZN"}
header = {"User-Agent": "my-app"}
news_html_table = {}
```

```python
# Prepare news table dictionary that conatins all the scrapped html

for tick in ticks:
    full_url = fetch_url + tick
    response = requests.get(full_url, headers=header)
    soup = BeautifulSoup(response.content, "html")
    news_html_table[tick] = soup.find(id="news-table")
```

```python
print(news_html_table)
```

```python
news_parsed_table = {}
```

```python
len("Today 07:15AM".split())
```

```python
for tick in ticks:
    table_rows = news_html_table[tick].find_all("tr")
    list_of_parsed_rows = []
    for index, row in enumerate(table_rows):
        title = row.a.text
        if len(row.td.text.split()) > 1:
            date = row.td.text.split()[0]
            timestamp = row.td.text.split()[1]
        parsed_row = {"Date":date, "Timestamp": timestamp, "Title":title}
        list_of_parsed_rows.append(parsed_row)
    news_parsed_table[tick] = pd.DataFrame.from_dict(list_of_parsed_rows)
        
```
