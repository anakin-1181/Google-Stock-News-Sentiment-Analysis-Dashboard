```python
import feedparser
import pandas as pd
from datetime import datetime
from urllib.parse import quote
```

```python
ticker = "NVDA"
company_name = "Nvidia"
query = f"{ticker} OR {company_name} when:7d"
encoded_query = quote(query)

url = f"https://news.google.com/rss/search?q={encoded_query}&hl=en-US&gl=US&ceid=US:en"
```

```python
feed = feedparser.parse(url)
feed
```

```python
df = pd.DataFrame.from_dict(feed.entries)
```

```python
df
```

```python
df = df.loc[:, ["published", "title"]]
df.columns = ["Date", "Title"]
df["Date"] = pd.to_datetime(df["Date"])
df["is_weekday"] = df["Date"].dt.day_name()
df["is_weekday"] = ((df["is_weekday"] != "Saturday") & (df["is_weekday"] != "Sunday"))
df["Date"] = df["Date"].dt.date

df = df.sort_values("Date")
df
```
